```python
import os
import re
import requests
from bs4 import BeautifulSoup
from markdownify import markdownify as md
from fake_useragent import UserAgent

# Function to extract text and links from an HTML element
def extract_text_links(element):
    for a in element.find_all('a', href=True):
        link = a['href']
        link_match = re.match(r'https://gdpr-info.eu/(art)-(\d+)-gdpr', link)
        if link_match:
            a.string = f'[[Article {link_match.group(2)}]]'
            print (a.string)
            a.unwrap()
    return element
    print (element) 
    # Feel free to remove print statements - I left them in as it's helpful
    # to understand what's being generated at different stages of the process.


# Function to download and save a page as a Markdown file
def download_page(url, filename):
    ua = UserAgent()
    headers = {'User-Agent': ua.random}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        header = soup.find('h1')
        print(header)
        content = soup.find('div', {'class': 'entry-content'})
        if content:
            content_text = extract_text_links(content)
            page = str(header) + str(content_text)
            print(page)
            with open(filename, 'w', encoding='utf-8') as md_file:
                md_file.write(md(str(header) + str(content_text), heading_style="ATX"))
            print(f'Saved {filename}')
        else:
            print(f"Failed to extract content from {url}")
    else:
        print(f"Failed to download {url}")

# Main script
base_url = 'https://gdpr-info.eu/'
article_url_pattern = base_url + 'art-%d-gdpr/'

# Download and save articles
for i in range(1, 100):  # Change the range based on the number of articles you want to download
    url = article_url_pattern % i
    filename = f'/data-transfer/gdpr/Article {i}.md'
    download_page(url, filename)

print('Completed')

```
